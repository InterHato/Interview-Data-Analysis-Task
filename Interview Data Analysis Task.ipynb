```python
totalorders = (df
     .reset_index()
     .set_index("eater_request_timestamp_utc")
     .groupby(["segment",pd.Grouper(freq='W')])["order_uuid"].agg('count')
     .astype(int)
     .reset_index())
totalorders
```

```python
segments = set(totalorders['segment'])
plt.figure()
for segment in segments:
     selected_data = totalorders.loc[totalorders['segment'] == segment]
     plt.plot(selected_data['eater_request_timestamp_utc'], selected_data['order_uuid'], label=segment)
     
plt.legend()
plt.show()
```

● Total orders with promotions

```python
df_finall = (df
     .reset_index()
     .set_index("eater_request_timestamp_utc")
     .groupby(["segment",'promo_uuid',pd.Grouper(freq='W')])["net_inflows_local"].agg('count')
             
     .astype(int)
     .reset_index())
df_finall#
#df_finall['segment','promo_uuid']
```

```python
finalll= df_finall.loc[df_finall['promo_uuid'] == '\\N']
finallll = finalll.reset_index()
finallll
```

```python
good= finallll.join(totalorders['order_uuid']) 
good
```

```python
good['ratio']= (totalorders['order_uuid']-good['net_inflows_local'])/totalorders['order_uuid']
good
```

```python
segments = set(totalorders['segment'])
plt.figure()
for segment in segments:
     selected_data = good.loc[good['segment'] == segment]
     plt.plot(selected_data['eater_request_timestamp_utc'], selected_data['ratio'], label=segment)
     
plt.legend()
plt.show()
```

Total GBs for orders where a promotion was used

```python
promotiontotal = (promotion
     .reset_index()
     .set_index("eater_request_timestamp_utc")
     .groupby(["segment",pd.Grouper(freq='W')])['gross_bookings_local'].sum()
             
     .astype(int)
     .reset_index())
promotiontotal
```

```python
segments = set(promotiontotal['segment'])
plt.figure()
for segment in segments:
     selected_data = promotiontotal.loc[promotiontotal['segment'] == segment]
     plt.plot(selected_data['eater_request_timestamp_utc'], selected_data['gross_bookings_local'], label=segment)
     
plt.legend()
plt.show()
```

Total promotion spend


```python
df
```

```python
promotiontotalspend = (df
     .reset_index()
     .set_index("eater_request_timestamp_utc")
     .groupby(["segment",pd.Grouper(freq='W')])['promo_spend_local'].sum()
             
     .astype(int)
     .reset_index())
promotiontotalspend
```

```python
segments = set(totalorders['segment'])
plt.figure()
for segment in segments:
     selected_data = promotiontotalspend.loc[promotiontotalspend['segment'] == segment]
     plt.plot(selected_data['eater_request_timestamp_utc'], selected_data['promo_spend_local'], label=segment)
     
plt.legend()
plt.show()
```

```python
promotiontotalspend['costperpromo']= promotiontotal['gross_bookings_local']-promotiontotalspend['promo_spend_local']
```

```python
segments = set(totalorders['segment'])
plt.figure()
for segment in segments:
     selected_data = promotiontotalspend.loc[promotiontotalspend['segment'] == segment]
     plt.plot(selected_data['eater_request_timestamp_utc'], selected_data['costperpromo'], label=segment)
     
plt.legend()
plt.show()
```

Average Basket Size

```python
df
```

```python
totalGBS = (df
     .reset_index()
     .set_index("eater_request_timestamp_utc")
     .groupby(["segment",pd.Grouper(freq='W')])['gross_bookings_local'].sum()
             
     .astype(int)
     .reset_index())
totalGBS
```

```python
totalGBS['AVERAGEBASKETSIZE']=  totalGBS['gross_bookings_local']/totalorders['order_uuid']
totalGBS
```

```python
segments = set(totalGBS['segment'])
plt.figure()
for segment in segments:
     selected_data = totalGBS.loc[totalGBS['segment'] == segment]
     plt.plot(selected_data['eater_request_timestamp_utc'], selected_data['AVERAGEBASKETSIZE'], label=segment)
     
plt.legend()
plt.show()
```
